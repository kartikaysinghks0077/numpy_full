```python
import numpy as np

# Vectors
vec1 = np.array([1, 2, 3])
vec2 = np.array([4, 5, 6])

# Dot Product
dot_product = np.dot(vec1, vec2)
print("Dot Product:", dot_product)

# Matrices
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])

# Matrix Multiplication
matrix_product = np.matmul(matrix1, matrix2)
print("Matrix Multiplication:\n", matrix_product)

# Inner and Outer Product
inner_product = np.inner(vec1, vec2)
print("Inner Product:", inner_product)

outer_product = np.outer(vec1, vec2)
print("Outer Product:\n", outer_product)

# Matrix Determinant
determinant = np.linalg.det(matrix1)
print("Determinant:", determinant)

# Matrix Inverse
inverse_matrix = np.linalg.inv(matrix1)
print("Inverse Matrix:\n", inverse_matrix)

# Solving Linear Equations
# Coefficient matrix
A = np.array([[3, 1], [1, 2]])
# Constant terms
b = np.array([9, 8])
# Solve for x in Ax = b
solution = np.linalg.solve(A, b)
print("Solution to the linear equations:", solution)

# Eigenvalues and Eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(matrix1)
print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)

# Matrix Norm
norm = np.linalg.norm(matrix1)
print("Matrix Norm:", norm)

# Singular Value Decomposition (SVD)
U, S, V = np.linalg.svd(matrix1)
print("U matrix:\n", U)
print("Singular values:", S)
print("V matrix:\n", V)
```
